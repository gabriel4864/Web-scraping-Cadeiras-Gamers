```python
import pandas as pd
import numpy as np

df = pd.read_excel("cadeiras.xlsx")
display(df)
```

```python
cores = [
"amarelo", "amarela", "azul", "branco", "branca", "cinza", "creme", "dourado", "laranja", "marrom", "preto", "rosa", "roxo", "verde", "vermelho"
"bege", "graphite", "geada", "uva", "grey", "yellow", "blue", "silver", "red", "black", "off white", "white", "green"
]

def extrair_cor(texto):
    texto = texto.lower()
    for cor in cores:
        if cor in texto:
            return cor
    return "Desconhecida"

df["cor_principal"] = df["marca"].apply(extrair_cor)
display(df["marca", "cor_principal"])

```
